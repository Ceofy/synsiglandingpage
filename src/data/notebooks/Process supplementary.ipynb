```python
import pandas as pd
```

```python
#Table 2
t2a = pd.read_csv('../originalData/Table S2.csv', header=2)

#Remove NaN rows
t2b = t2a.dropna(how='all')

#Remove NaN columns
t2c = t2b.dropna(axis=1, how='all')

#Change headers
header = [
    'Gene Symbol',
    'Accession',
    'Description',
    'SumCoverage',
    'SumNum Proteins',
    'SumNum Unique Peptides',
    'SumNum Peptides',
    'SumNum PSMs',
    'Score A2',
    'Coverage A2',
    'Num Peptides A2',
    'Num PSM A2',
    'Score A4',
    'Coverage A4',
    'Num Peptides A4',
    'Num PSM A4',
    'Num AAs',
    'MW kDa',
    'calc pI'
]
t2c.columns = header

#Sort
t2d = t2c.sort_values('Gene Symbol')

#Remove rows where gene symbol is NaN
t2e = t2d[t2d['Gene Symbol'].notna()]

#Remove unused columns
t2f = t2e.drop(['Accession', 'Description'], axis=1)

#Write to file
t2f.to_csv('../tables2.csv', index=False)
```

```python
#Table 3
t3a = pd.read_csv('./originalData/Table S3.csv', header=2)

#Remove NaN rows
t3b = t3a.dropna(how='all')

#Remove NaN columns
t3c = t3b.dropna(axis=1, how='all')

#Change headers
header = [
    'Gene Symbol',
    'Accession',
    'Description',
    'SumCoverage',
    'SumNum Proteins',
    'SumNum Unique Peptides',
    'SumNum Peptides',
    'SumNum PSMs',
    'Score A2',
    'Coverage A2',
    'Num Peptides A2',
    'Num PSM A2',
    'Score A4',
    'Coverage A4',
    'Num Peptides A4',
    'Num PSM A4',
    'Num AAs',
    'MW kDa',
    'calc pI'
]
t3a.columns = header

#Sort
t3d = t3a.sort_values('Gene Symbol')

#Remove rows where gene symbol is NaN
t3e = t3d[t3d['Gene Symbol'].notna()]

#Remove unused columns
t3f = t3e.drop(['Accession', 'Description'], axis=1)

#Write to file
t3f.to_csv('../tables3.csv', index=False)
```
