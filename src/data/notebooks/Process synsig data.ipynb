```python
import pandas as pd
```

```python
full = pd.read_csv('./update_web_table_full_db.csv', index_col=0)
full.genes = full.genes.astype(str)
```

```python
old_full = pd.read_csv('./update_web_table_desc.csv', index_col=0)[['genes', 'description']]
old_full.genes = old_full.genes.astype(str)
```

```python
new_full = pd.merge(full, old_full, on='genes', how='outer')
new_full['training'] = 'no'
```

```python
pos = pd.read_csv('./training_pos_ms_table.csv', index_col=0)
pos['training'] = 'pos'
```

```python
neg = pd.read_csv('./training_neg_ms_table.csv', index_col=0)
neg['training'] = 'neg'
```

```python
neg
```

```python
all_ = pd.concat([new_full, pos, neg])
all_.head()
```

```python
all_[['name', 'source']] = all_.description.str.split('[', expand=True)
all_new = all_.drop(['source', 'description'], axis=1)
all_new
```

```python
func = pd.read_csv('./all_synsig_functions.csv', index_col=0)
func
```

```python
func = func.drop('Total')
func = func.drop('description', axis=1)
func
```

```python
func.loc[func['genes'] == 'VAMP5']
```

```python
new = pd.merge(all_new, func, on='genes', how='outer')
```
